# Waffle Charts

```python
df_dsn = df_can.loc[['Denmark','Norway','Sweden'], :]
df_dsn
```

```python
# LANGKAH MEMBUAT WAFFLE #

#step1: menentukan proporsi setiap kategori sehubungan dgn totalnya

#compute the proportion
total_values = sum(df_dsn['Total'])
category_proportions = [(float(value) / total_values) for value in df_dsn['Total']]

#print out proportion
for i, proportion in enumerate(category_proportions):
    print(df_dsn.index.values[i] + ':' + str(proportion))
```

```python
#step2: menentukan ukuran keseluruhan dari waffle chart

width = 40   #width of chart
height = 10  #height of chart

total_num_tiles = width * height   #total number of tiles

print('Total number of tiles is', total_num_tiles)
```

```python
#step3: gunakan proporsi tiap kategori utk menentukan jml ubin masing2 kategori

#compute the number of tiles per category
tiles_per_category = [round(proportion * total_num_tiles) for proportion in category_proportions]

#print out number of tiles per category
for i, tiles in enumerate(tiles_per_category):
    print(df_dsn.index.values[i] + ':' + str(tiles))
```

```python
#step4: membuat matriks yg menyerupai waffel dan mengisinya

#initialize the waffle chart as an empty matrix
waffle_chart = np.zeros((height, width))

#define indices to loop through wafale chart
category_index = 0
tile_index = 0

#populate the waffle chart
for col in range(width):
    for row in range(height):
        tile_index += 1
        
#if the number of tiles populated for the current category is equal to its corresponding allocated tiles
        if tile_index > sum(tiles_per_category[0:category_index]):
    
#...proced to the next category
            category_index += 1

#set the class value to an integer
        waffle_chart[row, col] = category_index

print('Waffle chart populated!')
```

```python
#lihat tampilan matriks

waffle_chart
```

```python
#step5: petakan matriks waffle chart menjadi visual

#instiante a new figure object
fig = plt.figure()

#use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()
```

```python
#step6: percantik chart

#instantiate a new figure object
fig = plt.figure()

#use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()

#get the axis
ax = plt.gca()

#set minor ticks
ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
ax.set_yticks(np.arange(-.5, (width), 1), minor=True)

#add gridlines based on minor ticks
ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

plt.xticks([])
plt.yticks([])
```

```python
#step7: tambahkan legend

#instantiate a new figure object
fig = plt.figure()

#use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()

#get the axis
ax = plt.gca()

#set minor ticks
ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
ax.set_yticks(np.arange(-.5, (width), 1), minor=True)

#add gridlines based on minor ticks
ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

plt.xticks([])
plt.yticks([])

#compute cumulatuve sum of individual categories to match color shemes between chart and legend
values_cumsum = np.cumsum(df_dsn['Total'])
total_values = values_cumsum[len(values_cumsum) - 1]

#create legend
legend_handles = []
for i, category in enumerate(df_dsn.index.values):
    label_str = category + ' (' + str(df_dsn['Total'][i]) + ')'
    color_val = colormap(float(values_cumsum[i])/total_values)
    legend_handles.append(mpatches.Patch(color=color_val, label=label_str))

#add legend to chart
plt.legend(handles=legend_handles,
          loc='lower center',
          ncol=len(df_dsn.index.values),
          bbox_to_anchor=(0., -0.2, 0.95, -1)
          )
```

# Word Clouds

```python
#import package

from wordcloud import WordCloud, STOPWORDS
print('Wordcloud is installed and imported!')
```

```python
#open the file

alice_novel = open('alice_novel.txt', 'r').read()
print('File downloaded and saved!')
```

```python
#gunakan set utk menghapus stopword yg berlebihan

stopwords = set(STOPWORDS)
```

```python
#buat word cloud object dan generate word cloud
#buat word cloud hny dg 2000 kata pertama dlm novel

#instantiate a word cloud object
alice_wc = WordCloud(
    background_color=r'white',
    max_words=2000,
    stopwords=stopwords
)

#geneate the word cloud
alice_wc.generate(alice_novel)
```

```python
#visualisasi

#display the cloud
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#ubah ukuran cloud

fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords.add('said')  #add words to stopwords

#re-generate the word cloud
alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
alice_mask = np.array(Image.open('alice_mask.png'))

fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#gunakam word cloud pd gambar

#instantiate a word cloud object
alice_wc = WordCloud(background_color='white', max_words=2000, mask=alice_mask, stopwords=stopwords)

#generate the word cloud
alice_wc.generate(alice_novel)

#display the word cloud
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#total imigrasi dr 1980-2013

total_immigration = df_can['Total'].sum()
total_immigration
```

```python
max_words = 90
word_string = ''
for country in df_can.index.values:
    
    #chek if country's name is a single-word name
    if len(country.split(' ')) == 1:
        repeat_num_times = int(df_can.loc[country,'Total']/float(total_immigration)*max_words)
        word_string = word_string + ((country + ' ') * repeat_num_times)
        
#display the generated text
word_string
```

```python
#create the word cloud
wordcloud = WordCloud(background_color='white').generate(word_string)

print('Word cloud created!')
```

```python
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import seaborn as sns

print('Seaborn Installed and Imported!')
```

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))
```

```python
df_tot.index = map(float, df_tot.index)

df_tot.reset_index(inplace=True)

df_tot.columns = ['year','total']

df_tot.head()
```

```python
ax = sns.regplot(x='year', y='total', data=df_tot)
```

```python
#create df_countries dataframe
df_countries = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()

#create df_total by summing across three countries for each year
df_total = pd.DataFrame(df_countries.sum(axis=1))

#reset index in place
df_total.reset_index(inplace=True)

#rename columns
df_total.columns = ['year', 'total']

#change column year from string to int to create scatter plot
df_total['year'] = df_total['year'].astype(int)

#define figure size
plt.figure(figsize=(15, 10))

#define background style and font size
sns.set(font_scale=1.5)
sns.set_style('whitegrid')

#generate plot & add title & ases labels
ax = sns.regplot(x='year', y='total', data=df_total, color='green', marker='+', scatter_kws={'s': 200})
ax.set(xlabel='Year', ylabel='Total Immigration')
ax.set_title('Total Immigration from Denmark, Sweden, and Nowrway to Canada from 1980-2013')
```

```python
df_incidents = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Police_Department_Incidents_-_Previous_Year__2016_.csv')
print('Dataset downloaded and read into a pandas dataframe!')
```

```python
df_incidents.head()
```

```python
df_incidents.shape
```

```python
#get the first 100 crimes

limit = 100
df_incidents = df_incidents.iloc[0:limit, :]
```

```python
df_incidents.shape
```

```python
#visualisasikan kejahatan yg terjadi di san fransisco

latitude = 37.77
longitude = -122.42
```

```python
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)
sanfran_map
```

```python
#instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

#loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

#add incidents to map
sanfran_map.add_child(incidents)
```

```python
#instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

#loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

#add pop up text to each marker on the map
latitudes = list(df_incidents.Y)
longitudes = list(df_incidents.X)
labels = list(df_incidents.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sanfran_map)
    
#addfoliuments to map
sanfran_map.add_child(incidents)
```

```python
#instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

#loop through the 100 crimes and add each to the incidents feature group
for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='yellow',
        fill=True,
        popup=label,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(sanfran_map)

#show map
sanfran_map
```

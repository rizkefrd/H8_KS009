# Pandas Basics

```python
import numpy as np
import pandas as pd 
```

```python
df_can = pd.read_excel("Canada.xlsx",
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)

print('Data read into a pandas dataframe!')
```

```python
#menghapus kolom yg tdk perlu gunakan .drop()

df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
```

```python
#ganti nama kolom dg rename()

df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

```python
#menambahkan kolom

df_can['Total'] = df_can.sum(axis=1)
```

```python
#melihat byk objek null

df_can.isnull().sum()
```

# Area Plot

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

df_top5 = df_can.head()

df_top5 = df_top5[years].transpose()

df_top5.head()
```

```python
#utk menghasilkan unstacked plot, berikan stacked-false

df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area',
            stacked=False,
            figsize=(20, 10),
            )

plt.title('Immigrants Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
#ubah alpha 0.5

df_top5.plot(kind='area',
            alpha=0.25,
            stacked=False,
            figsize=(20, 10),
            )

plt.title('Immigrants Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

# Two Types of Ploting

```python
#plot dgn scripting layer

df_top5.plot(kind='area', alpha=0.35, figsize=(20, 10))

plt.title('ImmigrantsTrend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
#plot dgn artist layer

ax = df_top5.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('Immigrants Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
#gunakan scripting layer utk membuat stacked area plot dari 5 negara imigrasi ke kanada daro 1980-2013
#gunakan nilai transparan 0.45

df_least5 = df_can.tail(5)
df_least5 = df_least5[years].transpose()

df_least5.plot(kind='area',
              alpha=0.45,
              stacked=True,
              figsize=(20, 10),
              )

plt.title('Immigrants Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
#gunakan artist layer utk membuat stacked area plot dari 5 negara imigrasi ke kanada daro 1980-2013
#gunakan nilai transparan 0.55

ax = df_least5.plot(kind='area', alpha=0.55, figsize=(20, 10))

ax.set_title('Immigrants Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

# Pie Charts

```python
#step1: gunakan metode pandas groupby utk meringkas data imigrasi berdasarkan continent

#group contries by continents and apply sum() function

df_continents = df_can.groupby('Continent', axis=0).sum()

print(type(df_can.groupby('Continent', axis=0)))

df_continents.head()
```

```python
#plot data: dg passing kind='pie'

df_continents['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                           )

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal')

plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]



df_continents['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                            labels=None,
                            pctdistance=1.12,
                            colors=colors_list,
                            explode=explode_list
                           )

plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12)
plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')
plt.show()
```

# Scatter Plots

```python
#step1: get dataset

#use the sum() to get the total population per year
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

#change years to type int
df_tot.index = map(int, df_tot.index)

#reset index to put in back in as a column
df_tot.reset_index(inplace=True)

#rename column
df_tot.columns = ['years', 'total']

#view
df_tot.head()
```

```python
df_countries = df_can.loc[['Denmark','Norway','Sweden'],years].transpose()
df_total = pd.DataFrame(df_countries.sum(axis=1))
df_total.reset_index(inplace=True)
df_total.columns = ['year', 'total']
df_total['year'] = df_total['year'].astype(int)
df_total.head()
```

```python
df_total.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='blue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

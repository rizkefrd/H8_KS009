# Naive Bayes Classifier

```python
#defining dataset

weather = ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Rainy',
          'Overcast', 'Sunny', 'Sunny', 'Rainy', 'Sunny', 'Overcast',
          'Overcast', 'Rainy']

temp = ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild',
       'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild']

play = ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes',
       'Yes', 'Yes', 'Yes', 'Yes', 'No']
```

```python
#encoding features
#ubah string jd angka, misal overcast = 0, rainy = 1, sunny = 2 dikenal sbg label encoding

from sklearn import preprocessing

le = preprocessing.LabelEncoder()

weather_encoded = le.fit_transform(weather)

print(weather_encoded)
```

```python
#converting string labels into numbers

temp_encoded = le.fit_transform(temp)

label = le.fit_transform(play)

print(temp_encoded)
print(label)
```

```python
#combining wather and temp into single listof tuples

features = list(zip(weather_encoded, temp_encoded))

features
```

```python
#import gaussian naive bayes model
from sklearn.naive_bayes import GaussianNB

#vreate gaussian classifier
model = GaussianNB()

#train the model using the training sets
model.fit(features, label)

#predct
predicted = model.predict([[0,2]])
print("Predicted Value:", predicted)
```

# Decision Tree Classifier

## Decision Tree Classifiering in Scikit-Learn

```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split

from sklearn import metrics
```

```python
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin',
            'bmi', 'pedigree', 'age', 'label']

pima = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/diabetes.csv', header=None, names=col_names)
```

```python
pima.head()
```

```python
pima.info()
```

```python
numer = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree', 'label']

for col in numer:
    pima[col] = pd.to_numeric(pima[col], errors='coerce')
```

```python
pima.dropna(inplace=True)
```

```python
#split dataset in features and target variabels
feature_cols = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree']

X = pima[feature_cols]  #features
y = pima.label  #target variabels
```

```python
#split dataset into training set and test set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
#create decision tress classifier object
clf = DecisionTreeClassifier()

#train decision tree classifier
clf = clf.fit(X_train, y_train)

#predict
y_pred = clf.predict(X_test)
```

```python
#model accuracy
metrics.accuracy_score(y_test, y_pred)
```

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
#the label species
print(iris.target_names)

#the names
print(iris.feature_names)
```

```python
#top 5 records
print(iris.data[0:5])
```

```python
#buat data frame

import pandas as pd

data=pd.DataFrame({'sepal length': iris.data[:,0],
                  'sepal width': iris.data[:,1],
                  'petal length': iris.data[:,2],
                  'petal width': iris.data[:,3],
                  'species': iris.target
})

data.head()
```

```python
from sklearn.model_selection import train_test_split

X=data[['sepal length', 'sepal width', 'petal length', 'petal width']]
y=data['species'] #label

#split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) #70% train dan 30% test
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print("Accuracy:" , metrics.accuracy_score(y_test, y_pred))
```

```python
#top 5

print(cancer.data[0:5])
```

```python
#set target

print(cancer.target)
```

```python
#split data

from sklearn.model_selection import train_test_split

#70% training dan 30% test
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=109)
```

```python
#generating  model

from sklearn import svm

clf = svm.SVC(kernel='linear')

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
#evaluating model

from sklearn import metrics

print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

tingkat klasifikasi 96.49%, dianggap sbg akurasi yg sgt baik

```python
#precision
print("Precision:", metrics.precision_score(y_test, y_pred))

#recall
print("Recall:", metrics.recall_score(y_test, y_pred))
```

precision 98% dan recall 96% dianggap sbg nilai yg sgt baik

```python
train.Loan_Amount_Term=np.log(train.Loan_Amount_Term)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_cv, y_train, y_cv = train_test_split(X, y, test_size=0.2)
```

```python
pred_cv = model.predict(x_cv)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
print(accuracy_score(y_cv, pred_cv))
matrix=confusion_matrix(y_cv, pred_cv)
print(matrix)
```

```python
print(accuracy_score(y_cv, pred_cv1))
matrix1=confusion_matrix(y_cv, pred_cv1)
print(matrix1)
```
